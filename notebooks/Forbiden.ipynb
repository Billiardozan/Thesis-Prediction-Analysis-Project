```python
import pandas as pd
import numpy as np
import random
from datetime import timedelta, datetime

# Parameter untuk data
jenis_sayuran = ["Sawi Putih", "Cabe Chilli", "Cabe Kriting", "Cabe Rawit", "Kol", "Buncis", "Tomat"]
harga_petani_range = {
    "Sawi Putih": (500, 5000),
    "Cabe Chilli": (20000, 80000),
    "Cabe Kriting": (20000, 80000),
    "Cabe Rawit": (20000, 100000),
    "Kol": (2000, 8000),
    "Buncis": (2000, 8000),
    "Tomat": (1000, 8000),
}
musim = ["Kemarau", "Hujan"]

# Fungsi untuk menghasilkan data harga harian
def generate_harga_harian(start_date, end_date):
    data = []
    current_date = start_date
    id_counter = 1
    
    while current_date <= end_date:
        for sayur in jenis_sayuran:
            harga_petani = random.randint(*harga_petani_range[sayur])
            harga_pasar = harga_petani * random.uniform(1.1, 1.15)
            musim_harga = musim[current_date.month % 2]
            data.append([id_counter, sayur, harga_petani, current_date, musim_harga, round(harga_pasar, 2)])
            id_counter += 1
        current_date += timedelta(days=1)
    
    return pd.DataFrame(data, columns=["Id", "Jenis Sayuran", "Harga Sayuran per KG", "Tanggal Harga", "Musim", "Harga Pasar"])

# Fungsi untuk menghasilkan data transaksi penjualan harian
def generate_transaksi_harian(start_date, end_date, harga_harian_df):
    data = []
    current_date = start_date
    
    while current_date <= end_date:
        destinasi = ["Pasar Gudang Sukabumi", "Pasar Pasundan Sukabumi", "Pasar Bogor", "Pasar Induk"]
        random.shuffle(destinasi)
        destinasi = destinasi[:random.randint(2, 4)]
        
        for dest in destinasi:
            jumlah_sayur = random.randint(4, 6)
            jenis_terjual = random.sample(jenis_sayuran, jumlah_sayur)
            total_kirim = random.randint(2, 6) * 1000  # dalam kg
            biaya_kirim = total_kirim * 500
            
            harga_sayur = harga_harian_df[
                (harga_harian_df['Tanggal Harga'] == current_date) & 
                (harga_harian_df['Jenis Sayuran'].isin(jenis_terjual))
            ]
            
            total_harga_petani = harga_sayur['Harga Sayuran per KG'].sum() * total_kirim / len(jenis_terjual)
            total_harga_pasar = harga_sayur['Harga Pasar'].sum() * total_kirim / len(jenis_terjual)
            harga_bersih = total_harga_pasar - biaya_kirim
            
            data.append([
                current_date, 
                ", ".join(jenis_terjual), 
                total_kirim, 
                total_harga_petani, 
                total_harga_pasar, 
                biaya_kirim, 
                harga_bersih, 
                dest
            ])
        
        current_date += timedelta(days=1)
    
    return pd.DataFrame(data, columns=["Tanggal", "Jenis Sayuran", "Total Pengiriman (kg)", "Total Harga Petani", "Total Harga Pasar", "Biaya Kirim", "Harga Bersih", "Destinasi"])

# Periode data
start_date = datetime(2021, 1, 1)
end_date = datetime(2023, 12, 31)

# Generate data harga harian
harga_harian_df = generate_harga_harian(start_date, end_date)
# Generate data transaksi harian
transaksi_harian_df = generate_transaksi_harian(start_date, end_date, harga_harian_df)

# Simpan ke CSV
harga_harian_df.to_csv('harga_harian_sayuran.csv', index=False)
transaksi_harian_df.to_csv('transaksi_penjualan_harian.csv', index=False)


```

```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# Define date range
start_date = datetime(2021, 1, 1)
end_date = datetime(2023, 12, 31)
date_range = pd.date_range(start_date, end_date, freq='D')

# Define vegetable types and their price ranges
vegetables = {
    'Sawi Putih': (500, 5000),
    'Cabe Chilli': (20000, 80000),
    'Cabe Kriting': (20000, 80000),
    'Cabe Rawit': (20000, 80000),
    'Kol': (2000, 8000),
    'Buncis': (2000, 8000)
}

# Define seasons
def get_season(date):
    month = date.month
    if month in [11, 12, 1, 2, 3]:
        return 'Hujan'
    elif month in [4, 5, 6, 7, 8, 9, 10]:
        return 'Kemarau'
    else:
        return 'Pancaroba'

# Generate the data
data = []
id_counter = 1

for date in date_range:
    season = get_season(date)
    for vegetable, price_range in vegetables.items():
        # Adjust price based on season
        min_price, max_price = price_range
        if vegetable == 'Sawi Putih' and season != 'Pancaroba':
            continue
        if vegetable.startswith('Cabe') and season not in ['Pancaroba', 'Hujan']:
            continue
        if vegetable == 'Kol' and season != 'Kemarau':
            continue

        wholesale_price = random.uniform(min_price, max_price)
        market_price = wholesale_price * random.uniform(1.05, 1.10)

        data.append({
            'ID': id_counter,
            'Jenis Sayuran': vegetable,
            'Tanggal Harga': date,
            'Harga Pengepul per Kilogram': round(wholesale_price, 2),
            'Harga Pasar per Kilogram': round(market_price, 2)
        })

        id_counter += 1

# Create DataFrame
df = pd.DataFrame(data)

# Save to CSV
df.to_csv('harga_sayuran_2021_2023.csv', index=False)

print("Data generation complete. The data has been saved to 'harga_sayuran_2021_2023.csv'.")
```

```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# Data informasi sayuran dan harga
vegetables = {
    'Sawi Putih': (500, 5000),
    'Cabe Chilli': (20000, 80000),
    'Cabe Kriting': (20000, 80000),
    'Cabe Rawit': (20000, 80000),
    'Kol': (2000, 8000),
    'Buncis': (2000, 8000)
}

# Fungsi untuk mengenerate harga pasar
def generate_market_price(wholesale_price):
    return wholesale_price * random.uniform(1.05, 1.10)

# Membuat list untuk menyimpan data
data = []

# Mengenerate data dari tahun 2021 hingga 2023
start_date = datetime(2021, 1, 1)
end_date = datetime(2023, 12, 31)
date_range = pd.date_range(start_date, end_date)

# Membuat data harga perhari persayuran
id_counter = 1
for single_date in date_range:
    for vegetable, price_range in vegetables.items():
        wholesale_price = random.randint(price_range[0], price_range[1])
        market_price = generate_market_price(wholesale_price)
        
        data.append([id_counter, vegetable, single_date.date(), wholesale_price, market_price])
        id_counter += 1

# Membuat DataFrame
df = pd.DataFrame(data, columns=['ID', 'Jenis Sayuran', 'Tanggal Harga', 'Harga Pengepul per Kilogram', 'Harga Pasar per Kilogram'])

# Menampilkan contoh data
print(df.head())

# Menyimpan data ke file CSV
df.to_csv('harga_sayuran_2021_2023.csv', index=False)
```
